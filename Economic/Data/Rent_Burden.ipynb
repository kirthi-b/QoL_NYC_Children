```python
#table merge with dataset providing four-digit PUMA codes
df_puma = pd.read_excel("QoL_NYC_Children/Economic/Data/PUMA_Codes.xlsx")
df_puma.head()
```

```python
df_rentburden = pd.merge(df_puma, df_burden, on='Selected_Geographies', how='left')
df_rentburden.head()
```

```python
df_rentburden.to_csv('RentBurden.csv', index = False)
```
