```python
#loading file
df_socn1 = pd.read_excel("QoL_NYC_Children/Social/Data/Copy of soc_2019_acs5yr_cdta-1 (1).xlsx")
df_socn2 = pd.read_excel("QoL_NYC_Children/Social/Data/Copy of demo_2019_acs5yr_cdta-1.xlsx")
df_socn3 = pd.read_csv("QoL_NYC_Children/Social/Data/CorrectOne.csv")
```

```python
#filtering for important columns, here median income, splitting code and name, and create new dataframe
df_soc = pd.DataFrame(data={"Puma_Code" : list(df_socn1["GeogName"].str.split(" ", 1, expand = True).reset_index()[0]),
                            "Puma_Name" : list(df_socn1["GeogName"].str.split(" ", 1, expand = True).reset_index()[1]),
                            "Median_Income" : list(df_socn1['Rshp_ChE']),
                            "Pop_under_18" : list(df_socn2['PopU181E']),
                            "Pop_under_5" : list(df_socn2['PopU5E']),
                            "Pop_5to9" : list(df_socn2['Pop5t9E']),
                            "Pop_10to14" : list(df_socn2['Pop10t14E']),
                            "Pop_15to19" : list(df_socn2['Pop15t19E']),
                            "Hispanic_or_Latino" : list(df_socn2['Hsp1E']),
                            "White" : list(df_socn2['WtNHE']),
                            "Black_or_African_American" : list(df_socn2['BlNHE']),
                            "American_Indian_or_Alaska_Native" : list(df_socn2['AIANNHE']),
                            "Asian" : list(df_socn2['AsnNHE']),
                            "Native_Hawaiian_and_other_Pacific_Islander" : list(df_socn2['NHPINHE']),
                            "Other_Races" : list(df_socn2['OthNHE']),
                           }
                     )
#missing values from df_socn3, Total Population and Citizenship BUTBUTBUT (!!!!) the excle files have different amount of geographic areas
```

```python
df_soc.rename(columns={'Puma_Code': 'Code'}, inplace=True)
```

```python
#import dataset for equivalencies between PUMA Codes and Community District Codes
df_puma = pd.read_excel("QoL_NYC_Children/Economic/Data/PUMA_CD.xlsx")
```

```python
df_soc
```

```python
df_puma['PUMA'] = df_puma['PUMA'].astype(str)
df_puma['Code'] = df_puma['Code'].astype(str)
```

```python
#drop duplicates and null values
df_puma = df_puma.drop_duplicates(subset=['Code'])
df_puma = df_puma.reset_index(drop=True)
```

```python
#table merge with economics data with PUMA based on "Code"
df_social = pd.merge(df_puma, df_soc, on='Code', how='left')
df_social = df_social.dropna().reset_index(drop=True)
```

```python
df_social = df_social.dropna()
```
